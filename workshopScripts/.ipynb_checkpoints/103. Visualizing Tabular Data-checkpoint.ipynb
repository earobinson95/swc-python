```python
# import libraries
import numpy as np
import os
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# import data
data = np.loadtxt(fname = "../data/inflammation-01.csv", delimiter = ",")
```

```python
# plot all the data, color indicates the inflammation
image = matplotlib.pyplot.imshow(data)
matplotlib.pyplot.show()
```

```python
avg_inflammation = np.mean(data, axis = 0)
avg_plot = matplotlib.pyplot.plot(avg_inflammation)
matplotlib.pyplot.show()
```

```python
max_plot = matplotlib.pyplot.plot(numpy.max(data,axis = 0))
plt.show()
```

```python
minplot = plt.plot(np.min(data, axis = 0)) # minimum inflamation per day over all patients
plt.show()
```

```python
fig = plt.figure(figsize = (10.0, 3.0)) # dimensions in inches (rough estimate)
axes1 = fig.add_subplot(2, 2, (1,2)) # indicates row, col, place (by row)
axes2 = fig.add_subplot(2, 2, 3)
axes3 = fig.add_subplot(2, 2, 4)

# can adjust above... think in tuples fix.add_subplot(1, 3, (1,2))

axes1.set_ylabel("average")
axes1.set_xlabel("days")
axes1.plot(np.mean(data, axis = 0))

axes2.set_ylabel("max")
axes2.set_xlabel("days")
axes2.plot(np.max(data, axis = 0))

axes3.set_ylabel("min")
axes3.set_xlabel("days")
axes3.plot(np.min(data, axis = 0))
axes3.set_ylim(0,6) # adjust range and domain

fig.tight_layout()

plt.show()
```

```python
# create a plot showing the standard deviation (numpy.std) of the inflammation data for each day across all patients.
stdplot = plt.plot(np.std(data, axis = 0))
plt.xlabel("days")
plt.ylabel("standard deviation")
plt.title("Inflammation")
plt.show()
```
