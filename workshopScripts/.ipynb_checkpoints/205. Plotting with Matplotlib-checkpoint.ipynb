```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y,'ro') # use red dots
plt.show()
```

```python
import pandas_datareader as pdf
import datetime as dt
start = dt.datetime(1992,1,1)
end = dt.datetime.today()
df = pdr.data.get_data_fred(['HPIPONM226S', 'MSACSR'], start, end)
df.head()
```

```python
plt.plot(df)
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(df.HPIPONM226S, '--')
plt.title('Purchase Only House Price Index for the US')

plt.subplot(2,1,2)
plt.plot(df.MSACSR)
plt.title('Monthly Supply of Houses in the US')

plt.tight_layout()
plt.show()
```

```python
plt.figure(1, figsize = (16,4))
plt.plot(df.HPIPONM226S, '--')
plt.title('Purchase Only House Price Index for the US')

plt.figure(2, figsize = (8,4), facecolor=(0.6,0.6,0.6))
plt.plot(df.MSACSR)
plt.title('Monthly Supply of Houses in the US')

plt.tight_layout()
plt.show()
```

## Adding text to figures

```python
hpi = df.HPIPONM226S.loc['2005-01-01':'2015-01-01']
print(hpi)
```

```python
plt.figure(figsize=(16,6))
plt.plot(hpi)

plt.title('House Price Index for the US')
plt.grid(True)

axes = plt.gca()
axes.set_ylim([160,240])
axes.set_xlim([hpi.index[0], hpi.index[-1]])

plt.text(hpi.idxmin(), hpi.min()+3, "Min:" +str(hpi.min())+"\n"+hpi.idxmin().strftime('%Y-%m-%d'))
plt.text(hpi.idxmax(), hpi.max()+1, "Max:" +str(hpi.max())+"\n"+hpi.idxmax().strftime('%Y-%m-%d'))

plt.annotate(
    'When I bought my house\n' + str(hpi.loc['2005-12-01']),
    xy = (dt.datetime.strptime('2005-12-01', '%Y-%m-%d'), hpi.loc['2005-12-01']),
    xytext = (dt.datetime.strptime('2005-12-01', '%Y-%m-%d'), hpi.loc['2005-12-01']+10),
    arrowprops = dict(facecolor = 'black', shrink = 0.05, width = 1.0)
)

plt.annotate(
    'When I sold my house\n' + str(hpi.loc['2011-07-01']),
    xy = (dt.datetime.strptime('2011-07-01', '%Y-%m-%d'), hpi.loc['2011-07-01']),
    xytext = (dt.datetime.strptime('2011-07-01', '%Y-%m-%d'), hpi.loc['2011-07-01']-10),
    arrowprops = dict(facecolor = 'black', shrink = 0.05, width = 1.0, capstyle = "round", edgecolor = 'yellow')
)

plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(1239487)

# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure(1)

#linear
plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

#log
plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

#symmetric log
plt.subplot(2,2,3)
plt.plot(x,y)
plt.yscale('log')
plt.xscale('log')
plt.title('symlog')
plt.grid(True)

#logist
plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.tight_layout()
plt.show()
```
