```python
# import libraries
import numpy as np
import os
%matplotlib inline
import matplotlib.pyplot as plt
import glob
```

```python
print(glob.glob('../data/inflammation*.csv'))
```

```python
filenames = sorted(glob.glob('../data/inflammation*.csv'))
filenames = filenames[0:3]
for filename in filenames:
    print(filename)
    data = np.loadtxt(fname = filename, delimiter = ',')
    
    fig = plt.figure(figsize = (10.0, 3.0)) # dimensions in inches (rough estimate)
    axes1 = fig.add_subplot(1, 3, 1) # indicates row, col, place (by row)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)

    # can adjust above... think in tuples fix.add_subplot(1, 3, (1,2))

    axes1.set_ylabel("average")
    axes1.set_xlabel("days")
    axes1.plot(np.mean(data, axis = 0))

    axes2.set_ylabel("max")
    axes2.set_xlabel("days")
    axes2.plot(np.max(data, axis = 0))

    axes3.set_ylabel("min")
    axes3.set_xlabel("days")
    axes3.plot(np.min(data, axis = 0))

    fig.tight_layout()
    plt.show()
```

#### Plot the difference between the average inflammations reported in the first and second datasets (stored in inflammation-01.csv and inflammation-02.csv, correspondingly), i.e., the difference between the leftmost plots of the first two figures.

```python
# np.diff(patient_1_days_1_thru_5) 
data0 = np.loadtxt(fname = filenames[0], delimiter = ',')
data1 = np.loadtxt(fname = filenames[1], delimiter = ',')

fig = plt.figure(figsize = (10.0, 3.0))
plt.ylabel('Difference in average')
plt.plot(np.mean(data0, axis = 0) - np.mean(data1, axis = 0)) # calculate the difference in averages
fig.tight_layout()
plt.show()
```

```python
filenames = sorted(glob.glob('../data/inflammation*.csv'))
composite_data = np.zeros((60,40))
for filename in filenames:
    # sum each new file's data into composite_data as it's read
    data = np.loadtxt(fname = filename, delimiter = ',')
    composite_data = composite_data + data
# and then divide the composite data by number of samples
composite_data /= len(filenames)
# same as composite_data = composite_data / len(filenames)

# use pyplot to generate average, max and min for all patients:
fig = plt.figure(figsize = (10.0, 3.0)) # dimensions in inches (rough estimate)
axes1 = fig.add_subplot(1, 3, 1) # indicates row, col, place (by row)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

# can adjust above... think in tuples fix.add_subplot(1, 3, (1,2))

axes1.set_ylabel("average")
axes1.set_xlabel("days")
axes1.plot(np.mean(composite_data, axis = 0))

axes2.set_ylabel("max")
axes2.set_xlabel("days")
axes2.plot(np.max(composite_data, axis = 0))

axes3.set_ylabel("min")
axes3.set_xlabel("days")
axes3.plot(np.min(composite_data, axis = 0))

fig.tight_layout()
plt.show()
```
